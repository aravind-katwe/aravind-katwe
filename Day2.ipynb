### Question: Check if a number is positive, negative, or zero.

**Instructions:**
1. Assign an integer value to a variable named `number`.
2. Use an `if-elif-else` statement to determine if the `number` is:
   - Greater than 0 (positive)
   - Less than 0 (negative)
   - Equal to 0 (zero)
3. Print a message indicating whether the number is positive, negative, or zero.

```python
# Assign a number to a variable
number = 0  # You can change this number to test different cases

# Check if the number is positive, negative, or zero using if-elif-else
if number > 0:
  print(f"The number {number} is positive.")
elif number < 0:
  print(f"The number {number} is negative.")
else:
  print(f"The number {number} is zero.")
```
