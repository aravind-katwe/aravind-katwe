A **`for` loop** is used for iterating over a sequence (that is, a list, tuple, dictionary, set, or string) or other iterable objects. It executes a block of code for each item in the sequence. `for` loops are typically used when you know the number of iterations beforehand, or when you need to process each item in a collection.

Here's an example:

```python
# Example of a for loop
print("For loop example:")
for i in range(5):  # Iterates from 0 to 4
  print(i)
```

A **`while` loop** is used to execute a block of code repeatedly as long as a given condition is true. It's suitable when you don't know the exact number of iterations in advance, but you have a condition that will eventually become false to stop the loop. If the condition never becomes false, the loop will run indefinitely (an infinite loop).

Here's an example:

```python
# Example of a while loop
print("\nWhile loop example:")
count = 0
while count < 5: # Continues as long as 'count' is less than 5
  print(count)
  count += 1 # Increment count to eventually make the condition false
```

**Key Differences:**

*   **When to use:**
    *   `for` loops are generally used when you know the number of iterations or when you're iterating over a fixed collection.
    *   `while` loops are used when the number of iterations is unknown and depends on a condition being met.
*   **Control:**
    *   `for` loops iterate through elements of a sequence.
    *   `while` loops repeat based on a condition evaluating to `True`.
